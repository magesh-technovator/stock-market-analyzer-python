```python
import pandas as pd

df = pd.read_excel("overallData_latest.xlsx")
```

```python
data = df.drop(['Moving Averages', 'Technical Indicators',
       'Moving Averages Crossovers', 'symbol_url', 'Market Cap (Rs Cr.)',
       'P/E', 'Book Value (Rs)', 'Dividend (%)', 'Market Lot', 'Industry P/E',
       'EPS (TTM)', 'P/C', 'Price/Book', 'Dividend Yield.(%)',
       'Face Value (RS)', 'swot_url'], axis=1)
```

```python
data.columns
```

```python
data.rename(columns={"Unnamed: 0": "Company Names"})
```

```python
import ast

def get_unique_strenghts_and_count(swot):
    allSwot = {}
    
    aggregate = list(swot)
    
    for row in aggregate:
        try:
            if type(row) == str and not row.endswith("found") and row != "Data Not Available":
                points = ast.literal_eval(row)
                
                for point in points:
                    if point not in allSwot.keys():
                        allSwot[point] = 1
                    else:
                        allSwot[point] += 1
                    
        except:
            print(row)
            raise Exception
    
    return allSwot
```

```python
swot_types = ["Strengths", "Weaknesses", "Opportunities", "Threats"]
analyser = {}
for type_ in swot_types:
    analyser[type_] = get_unique_strenghts_and_count(data[type_])
    print(type_, len(analyser[type_]))
```

```python
def analyze_swot(swot):
    tempdf = pd.DataFrame()
    tempdf["Points"] = swot.keys()
    tempdf["Counts"] = swot.values()
    return tempdf
```

```python
writer = pd.ExcelWriter('swot_analysis.xlsx', engine='xlsxwriter')

for type_ in swot_types:
    new_df = analyze_swot(analyser[type_])
    new_df.to_excel(writer, sheet_name=type_, index=False)
    
writer.save()
```
